```python
import base64
import requests
import os
import json
import random
# OpenAI API Key
api_key = "api_key"

headers = {
  "Content-Type": "application/json",
  "Authorization": f"Bearer {api_key}"
}

files=["gpt4-turbo_clean","gpt-4o_clean","gemini_clean"]
for file_name in files:
  with open(f"{file_name}.txt","r") as file:
    while True:
      lines = [file.readline() for _ in range(20)]
      text = ''.join([line for line in lines if line])
      if not text:
        break
      payload = {
        "model": "gpt-4o",
        "messages": [
          {
            "role": "user",
            "content": [
              {
                "type": "text",
                "text": f"Your job is to label the attributes in the paragraphs I've given you according to the categories I give you. Note that you should output the same format (in json format) as I give you. Do not change the format! When there are multiple options separated by commas in an attribute, you should separate them in the form of list！ Here is the categories: {categories}, And here is the paragraph:{text} "
              }
            ]
          }
        ],
        "max_tokens": 1024,
      }

      response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)
      text_after = response.json()["choices"][0]["message"]["content"]

      with open(f"{file_name}_labelled.txt","a+") as f:
        f.write(text_after+"\n")
      print(text_after)
      print("")
```
